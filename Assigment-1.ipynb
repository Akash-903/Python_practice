```python
#importing the libraries
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
#load the data
df=pd.read_csv("heart.csv.csv")
df.head()
```

```python
df.describe()
```

```python
#checking the missing values
df.isnull().sum()
```

```python
#loading input and output variables
x=df.drop("target",axis=1)
y=df["target"]
x
```

```python
#split the data into train and test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
#train the model
from sklearn.linear_model import LogisticRegression
# load the model
lg=LogisticRegression()
# fit the model
train_model = lg.fit(x_train,y_train)
#make predictions
y_pred=train_model.predict(x_test)
y_pred
```

```python
#find the accuarcy
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y_test,y_pred)
print("Accuracy of logistic regression model is:",accuracy)
```

```python
#check loss function
from sklearn.metrics import log_loss
logloss= log_loss(y_test,y_pred)
print("Log loss of Logistic Regression model is:",logloss)
```

```python
#combined actual vs predicted
comparsion = pd.DataFrame({"actual:":y_test, "predicted":y_pred})
comparsion
```

```python
#find the confusion matrix
from sklearn.metrics import confusion_matrix
cm= confusion_matrix(y_test,y_pred)
cm
```

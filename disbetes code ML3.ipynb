```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('diabetes.csv')
df.head()
```

```python
# check the missing value
df.isnull().sum()
```

```python
df.columns = df.columns.str.strip().str.lower()
x = df.drop("outcome", axis=1)
y = df["outcome"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42)
```

```python
#train the model
from sklearn.linear_model import LogisticRegression
# load the model
lg=LogisticRegression()
# fit the model
train_model = lg.fit(x_train,y_train)
#make predictions
y_pred=train_model.predict(x_test)
y_pred
```

```python
# find accuracy
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y_test,y_pred)
print("accuracy of logistic regression model is:",
accuracy)
```

```python
# check loss function
from sklearn.metrics import log_loss
logloss=log_loss(y_test,y_pred)
print("log loss of logistic regression model is:",logloss)
```

```python
# comined actual vs predicted
comparsion=pd.DataFrame({"actual:":y_test,"predicted":y_pred})
comparsion
```

```python
#find the confusion matrix
from sklearn.metrics import confusion_matrix
cm= confusion_matrix(y_test,y_pred)
cm
```

```python
from scipy import stats

# split glucose based on outcome
Glucose_pos = df[df['outcome'] == 1]['glucose']
Glucose_neg = df[df['outcome'] == 0]['glucose']

# independent t-test
t_stat, p_val = stats.ttest_ind(Glucose_pos, Glucose_neg)

print("Independent T-Test statistic:", t_stat)
print("P-value:", p_val)

if p_val < 0.05:
    print("Reject null hypothesis")
else:
    print("Fail to reject null hypothesis")
```

# T-Test
**Definition:** T-Test is used to compare means when the popualtion variance is unknown or the sample size is small

--

## Independent T-Test
**Example:** Compare **Glucose** of **Positive vs Negative** person.<br>
**Hypotheses:**<br>
-H0: Glucose distribution of positive = Glucose distribution of negative<br>
-H1: Glucose distribution of positive is not equal Glucose distribution of negative

```python
from scipy import stats

# split glucose based on outcome
pregnancies_pos = df[df['outcome'] == 1]['pregnancies']
pregnancies_neg = df[df['outcome'] == 0]['pregnancies']

# independent t-test
t_stat, p_val = stats.ttest_ind(pregnancies_pos, pregnancies_neg)

print("Independent T-Test statistic:", t_stat)
print("P-value:", p_val)

if p_val < 0.05:
    print("Reject null hypothesis")
else:
    print("Fail to reject null hypothesis")
```
